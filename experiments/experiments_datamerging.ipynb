```python
import pandas as pd
import dedupe
```

```python
data = pd.read_csv('../dblp/dblp-1.csv')
data.info()
```

```python
data['title_end'] = data['ptitle'].apply(lambda x : x[-1])
data['pipe'] = data['ptitle'].apply(lambda x : '|' in x)
data.loc[(data['title_end'] != '.') & (data['pipe'] == False) & (data['ptitle'].str.len() > 40)]
```

```python
fields = [
    {'field': 'pauthor', 'type': 'String'},
    {'field': 'ptitle', 'type': 'String'}]

deduper = dedupe.Dedupe(fields)

deduper.prepare_training(data.T.to_dict())

dedupe.console_label(deduper)
```
